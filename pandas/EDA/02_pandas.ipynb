# hOW TO RENAME COLUMNS

```python
df = pd.DataFrame({'A col':[1,2,3,45,56,], 'B col':[4,5,6,56,678]})
df
```

```python
df.rename(columns={'A col':'col_a', 'B col':'col_b'}, inplace=True)
df
```

```python
#rename column names
df.columns=['col_aa','col_bb']
df
```

```python
# to replace any characters
df.columns=df.columns.str.replace('_','*')
df
```

```python
#adding prefixes to columns
df = df.add_prefix('baba*')
df
```

```python
# Adding suffix to column names
df = df.add_suffix('*khan')
df
```

```python
df.columns = ['cal_a', 'cal_b']
df
```

## covert strings to number

```python
df = pd.DataFrame({'col_A':['1','2','3','4','5','6',], 'col_B':['7','8','9','10','11','12']})
df
```

```python
df.dtypes
```

```python
df.astype({'col_A': 'int64', 'col_B': 'int64'}).dtypes
```

```python
pd.to_numeric(df['col_A'],errors='coerce')
pd.to_numeric(df['col_B'],errors='coerce')
```
